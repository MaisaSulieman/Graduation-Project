```python
def merge_photos(dir):
    image_files=[filename for filename in os.listdir(dir) if filename.endswith(".jpg")]
    
    image_files.sort()
    
    for i in range(0, len(image_files), 4):
        images=[Image.open(os.path.join(dir, filename)) for filename in image_files[i:i+4]]
        
        w,h=images[0].size
        images=[image.resize((w,h), Image.Resampling.LANCZOS) for image in images]
        merged_image = Image.new("RGB", (2 * w, 2 * h))
        
        merged_image.paste(images[0], (0, 0))
        merged_image.paste(images[1], (w, 0))
        merged_image.paste(images[2], (0, h))
        merged_image.paste(images[3], (w, h))
        
        merged_image.save(f"merged_image_{i//4}.jpg") 
```

```python
directory='filepath'
merge_photos(directory)
```
